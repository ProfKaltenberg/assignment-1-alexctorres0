```python
##I was having trouble with the foursquare API so I decided to change the concept of my project. I was also convinced to try something more interesting/useful in the field of economics after hearing many of the Pace Econ alumns use this project as a beginning step into doing real world work.
##the folowing changes I made are:
##1) I took out Foursquare and the related restaurant variables from the project and decided to only use theBEA and Census APIs. I swithced because in the proposal you warned that the Foursquare API might be too time consuming and you were rihgt. the data given was inconsistent so i decided to just usenthe other 2 APIs since i knew theyd  be more reliable. 

##2) I changed my question to something related to if a big city wages actually keep up with big city prices. I chnaged my question to a better and stromger academic question related to my field.
 
##3) my variables changed as well. I changed the average restaurant rating to nominal median household income, which I got from the Census API. I kept the pop and RPP variables. I also changed my calculated variable from resturant density to real purchasing power. this varible is better for my specific question and more economics related. To turn nominal pay into real pay, I divide the raw median income by the living costs percentage. 

##4) I also had to change my regtression model as a result. Before, RPP, which was the Independent variable, guessed restaurant rating, which was the dependent variable. RPP and Population (independent variables ) can now be used to predict median household income (dependent variable).
```

```python
import requests
import pandas as pd

##API KEYS
BEA_KEY = "8F024E50-295D-456C-8EF1-C7695242F328"
CENSUS_KEY = "f36641aa3e79a6f5ed7f6b0cc6c6d62acca9279d"
##the 15 cities 
cities = [ "New York, NY", "San Francisco, CA", "Boston, MA", "Washington, DC", "Seattle, WA", "Miami, FL", "Chicago, IL", "Denver, CO", "Austin, TX", "Atlanta, GA", "Phoenix, AZ", "Nashville, TN", "Houston, TX", "Kansas City, MO", "Memphis, TN"]

##the search terms for each city according to the way they list it on the BEA and Census sites  
search_terms = ["New York-Newark", "San Francisco-Oakland", "Boston-Cambridge", "Washington-Arlington", "Seattle-Tacoma", "Miami-Fort", "Chicago-Naperville", "Denver-Aurora", "Austin-Round", "Atlanta-Sandy", "Phoenix-Mesa", "Nashville-Davidson", "Houston", "Kansas City", "Memphis"]

##empty lists to hold our API data
city_names = []
rpp_scores = []
populations = []
median_incomes = []

for city in cities:##takes the city names and drop them into the first empty list
    city_names.append(city)

##the API url getting 2021 RPP for metro areas
bea_url = f"https://apps.bea.gov/api/data/?UserID={BEA_KEY}&method=GetData&datasetname=Regional&TableName=MARPP&LineCode=1&Year=2021&GeoFips=MSA&ResultFormat=json"
bea_response = requests.get(bea_url)
bea_data = bea_response.json()
bea_results = bea_data['BEAAPI']['Results']['Data']##get the list of city data rows only


for city in search_terms:
    city_score = None ##starts by assuming no score for this city yet
    for row in bea_results: ##looks at every single row of data the government gave us
        if city in row['GeoName']: ##if the city matches the government's geographic name, save the cost of living number for that city 
            city_score = row['DataValue']
    rpp_scores.append(city_score) ##add the number we found to the final list


##B01003_001E is population and B19013_001E is median household income 
census_url = f"https://api.census.gov/data/2021/acs/acs5?get=NAME,B01003_001E,B19013_001E&for=metropolitan%20statistical%20area/micropolitan%20statistical%20area:*&key={CENSUS_KEY}"
census_response = requests.get(census_url)
census_data = census_response.json()


for city in search_terms:## loops through the 15 specific city search terms
    ##assume missing numbers for both 
    city_pop = None
    city_inc = None
      
    for row in census_data[1:]: ##skips header
        metro_stat_area_name = row[0]##Census keeps the city name in the first column (0)
        if city in metro_stat_area_name:
            ##save the population and income numbers
            city_pop = row[1]
            city_inc = row[2]
            
    ##add the numbers to the final lists
    populations.append(city_pop)
    median_incomes.append(city_inc)

##put lists into individual dataframes
df_cities = pd.DataFrame(city_names, columns=['City'])
df_rpp = pd.DataFrame(rpp_scores, columns=['Cost_Of_Living_RPP'])
df_pop = pd.DataFrame(populations, columns=['Population'])
df_inc = pd.DataFrame(median_incomes, columns=['Median_Household_Income'])

##turns the API strings to numbers so we can do math on them
##used the following help: https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html
df_rpp['Cost_Of_Living_RPP'] = pd.to_numeric(df_rpp['Cost_Of_Living_RPP'])
df_pop['Population'] = pd.to_numeric(df_pop['Population'])
df_inc['Median_Household_Income'] = pd.to_numeric(df_inc['Median_Household_Income'])

##puts all DataFrames together side by side
final_df = pd.concat([df_cities, df_rpp, df_pop, df_inc], axis=1)

##new variable: calculates what a city's income is actually worth by adjusting for the cost of living
final_df['Real_Purchasing_Power'] = final_df['Median_Household_Income'] / (final_df['Cost_Of_Living_RPP'] / 100)

##sets decimals so it is in currency from
final_df['Real_Purchasing_Power'] = final_df['Real_Purchasing_Power'].round(2)

##export final csv
final_df.to_csv('final_dataset.csv', index=False)
print("All 15 cities saved as 'final_dataset.csv'")



##Proposed regression analysis:
##for this dataset, I will run a multiple linear regression using the Regional Price Parity (RPP) Index and Population as the independent variables (X1 and X2), and the Median Household Income as the dependent variable (Y) 
##the goal of this regression is to see if a city's wages really keep up with its prices by looking at the cost of living gap. This regression will show whether a higher cost of living leads to an equal rise in nominal wages by looking at the link between RPP and income and using population as a control variable for the size of the labor market.
```
